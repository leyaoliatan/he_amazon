# Initial test build a co-review network based on the data

```python
review_path = '../data/filtered_reviews.csv'
filtered_review_df = pd.read_csv(review_path)
product_path = '../data/UCSD_product_level_data.csv' #this is about 20M
label_path = '../data/product_level_data_without_img_feats.csv'
product_df = pd.read_csv(product_path)
label_df = pd.read_csv(label_path)
```

```python
#save label txt
with open('../data/network/ProductLabel.txt', 'w') as f:
    for index, row in label_df.iterrows():
        f.write(f"{int(row['product_ID'])} {int(row['fake'])}\n")
```

```python
# Group reviews by product_ID to find reviewers for each product
product_reviewers = filtered_review_df.groupby('product_ID')['reviewer_ID'].apply(list).reset_index()

# File to save the edges
output_file = '../data/network/Network.txt'

# Generate edges directly to file
with open(output_file, 'w') as f:
    edge_count = 0
    for _, row in product_reviewers.iterrows():
        reviewers = row['reviewer_ID']
        # Create edges between all pairs of reviewers for this product
        for i in range(len(reviewers)):
            for j in range(i+1, len(reviewers)):
                f.write(f"{reviewers[i]} {reviewers[j]}\n")
                edge_count += 1
        
        # Optional: Print progress for large datasets
        if _ % 100 == 0:
            print(f"Processed {_} products...", end='\r', flush=True)

print(f"Edge list created with {edge_count} edges")
print(f"Saved to {output_file}")

# If you want some basic statistics without loading the full graph
unique_reviewers = len(filtered_review_df['reviewer_ID'].unique())
print(f"Number of unique reviewers (nodes): {unique_reviewers}")
print(f"Average degree: {2*edge_count/unique_reviewers:.2f}")
```
